## Data Visualization

```python
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for col in  columns:
    plt.figure(figsize=(20,4))
    sns.countplot(dataset[col])
    plt.show()
```

```python
sns.pairplot(dataset, hue='species', size=3)
plt.show()
```

### Heatmap correlation

```python
fig = plt.figure(figsize = (12,10))
sns.heatmap(dataset.corr(), cmap='Blues', annot = True)
```

```python
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for col in columns:
    plt.figure(figsize=(10,6))
    sns.swarmplot(x=dataset['species'],y= dataset[col],data=dataset,palette='coolwarm')
    plt.title(f'Species vs {col}')
```

### Histograms

```python
fig = plt.figure(figsize=(20, 12))
dataset.hist(ax = fig.gca(), color="green", edgecolor="black")
plt.show()
```
